```python
"""
CONTINUOUS PREDICTION SYSTEM
Optimized for real-time, continuous predictions in a loop

Features:
- Predictions run every N seconds (configurable)
- Live progress bar
- Real-time price updates
- Continuous signal generation
- Auto-saves predictions periodically
"""

import asyncio
import json
import ssl
import websockets
import requests
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from collections import deque
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# CONFIGURATION - ADJUST THESE
# ============================================================================

ACCESS_TOKEN = 'eyJ0eXAiOiJKV1QiLCJrZXlfaWQiOiJza192MS4wIiwiYWxnIjoiSFMyNTYifQ.eyJzdWIiOiI3QkJFUzMiLCJqdGkiOiI2OTQ4ZDc2ZmMxMTIxNTM2NjJhOThiOGIiLCJpc011bHRpQ2xpZW50IjpmYWxzZSwiaXNQbHVzUGxhbiI6ZmFsc2UsImlhdCI6MTc2NjM4MTQyMywiaXNzIjoidWRhcGktZ2F0ZXdheS1zZXJ2aWNlIiwiZXhwIjoxNzY2NDQwODAwfQ.gHYPpYy4rgPsMY0Xr22QlICB7fTc03VSb7H6O8KbUgU'

# Prediction frequency (in seconds)
PREDICTION_INTERVAL = 10  # Run predictions every 10 seconds

# Status display frequency (in seconds)
STATUS_INTERVAL = 30  # Show status every 30 seconds

# Auto-save frequency (in seconds)
AUTOSAVE_INTERVAL = 300  # Save predictions every 5 minutes

# Minimum candles needed for predictions
MIN_CANDLES = 20  # Reduced for faster startup

INSTRUMENTS = {
    'NIFTY50': 'NSE_INDEX|Nifty 50',
    'BANKNIFTY': 'NSE_INDEX|Nifty Bank',
    'NIFTYAUTO': 'NSE_INDEX|Nifty Auto',
    'NIFTYIT': 'NSE_INDEX|Nifty IT'
}

# ============================================================================
# ENHANCED CONTINUOUS TRADING SYSTEM
# ============================================================================

class ContinuousTradingSystem:
    """Optimized system for continuous predictions"""

    def __init__(self):
        self.data = {inst: deque(maxlen=200) for inst in INSTRUMENTS.keys()}
        self.predictions = []
        self.models = self._load_models()

        # Timing trackers
        self.last_prediction_time = None
        self.last_status_time = None
        self.last_autosave_time = None
        self.start_time = None
        self.message_count = 0
        self.prediction_count = 0

        # Performance tracking
        self.prediction_times = []

    def _load_models(self):
        """Load ML models"""
        models = {}

        try:
            from tensorflow.keras.models import load_model
            models['LSTM'] = load_model('yourmodel.keras')
            models['CNN_LSTM'] = load_model('yourmodel1.keras')
            models['ATTENTION'] = load_model('yourmodel3.keras')
            print("✓ Deep learning models loaded")
        except Exception as e:
            print(f"⚠️ DL models not loaded: {e}")

        try:
            import joblib
            models['RF'] = joblib.load('RF_NiftyAuto_1m.joblib')
            print("✓ Random Forest loaded")
        except Exception as e:
            print(f"⚠️ RF model not loaded: {e}")

        return models

    async def run(self, duration_minutes=60):
        """Run continuous prediction system"""

        print("\n" + "="*80)
        print("🔄 CONTINUOUS PREDICTION SYSTEM")
        print("="*80)
        print(f"⏱️  Duration: {duration_minutes} minutes")
        print(f"🔮 Prediction interval: Every {PREDICTION_INTERVAL} seconds")
        print(f"📊 Status updates: Every {STATUS_INTERVAL} seconds")
        print(f"💾 Auto-save: Every {AUTOSAVE_INTERVAL} seconds")
        print(f"📈 Minimum candles: {MIN_CANDLES}")
        print("="*80 + "\n")

        # Get WebSocket connection
        ws_url = self._get_auth()
        if not ws_url:
            print("❌ Authorization failed!")
            print("\n💡 TIP: Your access token may have expired.")
            print("   Run the token generation code to get a fresh token.")
            return

        print("✓ Authorization successful")

        # Connect
        ssl_context = ssl.create_default_context()
        ssl_context.check_hostname = False
        ssl_context.verify_mode = ssl.CERT_NONE

        async with websockets.connect(ws_url, ssl=ssl_context) as ws:
            print("✓ WebSocket connected")

            # Subscribe
            await self._subscribe(ws)
            print("✓ Subscribed to instruments\n")
            print("📡 Receiving live data...\n")

            self.start_time = datetime.now()
            self.last_prediction_time = datetime.now() - timedelta(seconds=PREDICTION_INTERVAL)
            self.last_status_time = datetime.now()
            self.last_autosave_time = datetime.now()

            # Main loop
            end_time = self.start_time + timedelta(minutes=duration_minutes)

            while datetime.now() < end_time:
                try:
                    # Receive data with short timeout for responsive loop
                    msg = await asyncio.wait_for(ws.recv(), timeout=0.5)
                    data = self._decode(msg)

                    if data.get('type') == 'live_feed':
                        self._update_data(data)
                        self.message_count += 1

                    # Check if it's time for predictions
                    if self._should_run_predictions():
                        await self._run_predictions()

                    # Check if it's time for status update
                    if self._should_show_status():
                        self._show_status()

                    # Check if it's time for auto-save
                    if self._should_autosave():
                        self._autosave()

                except asyncio.TimeoutError:
                    # No message received, but check if we should do periodic tasks
                    if self._should_run_predictions():
                        await self._run_predictions()

                    if self._should_show_status():
                        self._show_status()

                    if self._should_autosave():
                        self._autosave()

                    continue

                except Exception as e:
                    print(f"❌ Error in main loop: {e}")
                    continue

            print("\n" + "="*80)
            print("✅ TRADING SESSION COMPLETED")
            print("="*80)
            self._print_final_summary()

    def _should_run_predictions(self):
        """Check if it's time to run predictions"""
        if self.last_prediction_time is None:
            return True

        elapsed = (datetime.now() - self.last_prediction_time).total_seconds()
        return elapsed >= PREDICTION_INTERVAL

    def _should_show_status(self):
        """Check if it's time to show status"""
        elapsed = (datetime.now() - self.last_status_time).total_seconds()
        return elapsed >= STATUS_INTERVAL

    def _should_autosave(self):
        """Check if it's time to auto-save"""
        if not self.predictions:
            return False
        elapsed = (datetime.now() - self.last_autosave_time).total_seconds()
        return elapsed >= AUTOSAVE_INTERVAL

    async def _run_predictions(self):
        """Run predictions for all instruments"""
        pred_start = datetime.now()
        new_predictions = 0

        for inst in INSTRUMENTS.keys():
            candles = list(self.data[inst])

            if len(candles) >= MIN_CANDLES:
                df = pd.DataFrame(candles)
                df['time'] = pd.to_datetime(df['time'])
                df.set_index('time', inplace=True)

                # Calculate indicators
                df['SMA_10'] = df['close'].rolling(10).mean()
                df['SMA_20'] = df['close'].rolling(20).mean()

                # Calculate RSI
                delta = df['close'].diff()
                gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
                loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
                rs = gain / (loss + 1e-10)
                df['RSI'] = 100 - (100 / (1 + rs))

                current = df['ltp'].iloc[-1]
                sma10 = df['SMA_10'].iloc[-1]
                sma20 = df['SMA_20'].iloc[-1]
                rsi = df['RSI'].iloc[-1]

                # Predict with models
                pred = self._predict_with_models(df, inst)

                # Determine signal
                signal = 'HOLD'
                if pd.notna(sma10) and pd.notna(sma20) and pd.notna(rsi):
                    if sma10 > sma20 and rsi < 70:
                        signal = 'BUY'
                    elif sma10 < sma20 and rsi > 30:
                        signal = 'SELL'

                # Store prediction
                self.predictions.append({
                    'time': datetime.now(),
                    'instrument': inst,
                    'price': current,
                    'signal': signal,
                    'RSI': rsi,
                    'SMA_10': sma10,
                    'SMA_20': sma20,
                    **pred
                })

                new_predictions += 1

        self.last_prediction_time = datetime.now()
        self.prediction_count += new_predictions

        # Track prediction performance
        pred_time = (datetime.now() - pred_start).total_seconds()
        self.prediction_times.append(pred_time)

        if new_predictions > 0:
            print(f"🔮 [{datetime.now().strftime('%H:%M:%S')}] "
                  f"Generated {new_predictions} predictions in {pred_time:.2f}s")

    def _predict_with_models(self, df, instrument):
        """Make predictions using loaded models"""
        predictions = {}

        try:
            # Prepare features
            feature_cols = ['open', 'high', 'low', 'close', 'ltp']
            features = df[feature_cols].tail(MIN_CANDLES).values

            # Normalize
            features = (features - features.mean(axis=0)) / (features.std(axis=0) + 1e-8)

            # LSTM models
            if 'LSTM' in self.models:
                X = features.reshape(1, MIN_CANDLES, len(feature_cols))
                pred = float(self.models['LSTM'].predict(X, verbose=0)[0][0])
                predictions['LSTM'] = pred

            if 'CNN_LSTM' in self.models:
                X = features.reshape(1, MIN_CANDLES, len(feature_cols))
                pred = float(self.models['CNN_LSTM'].predict(X, verbose=0)[0][0])
                predictions['CNN_LSTM'] = pred

            if 'ATTENTION' in self.models:
                X = features.reshape(1, MIN_CANDLES, len(feature_cols))
                pred = float(self.models['ATTENTION'].predict(X, verbose=0)[0][0])
                predictions['ATTENTION'] = pred

            # Random Forest
            if 'RF' in self.models:
                X = features.flatten().reshape(1, -1)
                pred = float(self.models['RF'].predict(X)[0])
                predictions['RF'] = pred

            # Ensemble
            if predictions:
                predictions['ENSEMBLE'] = np.mean(list(predictions.values()))

        except Exception as e:
            print(f"⚠️ Prediction error for {instrument}: {e}")

        return predictions

    def _show_status(self):
        """Show current status"""
        elapsed = (datetime.now() - self.start_time).total_seconds()

        print("\n" + "─"*80)
        print(f"⏱️  Runtime: {int(elapsed//60)}m {int(elapsed%60)}s | "
              f"Messages: {self.message_count} | "
              f"Predictions: {self.prediction_count}")
        print("─"*80)

        for inst in INSTRUMENTS.keys():
            count = len(self.data[inst])
            status = "✓" if count >= MIN_CANDLES else "⏳"

            if self.data[inst]:
                latest = list(self.data[inst])[-1]
                price = latest['ltp']

                # Get latest prediction if available
                inst_preds = [p for p in self.predictions if p['instrument'] == inst]
                if inst_preds:
                    latest_pred = inst_preds[-1]
                    signal = latest_pred.get('signal', 'HOLD')
                    ensemble = latest_pred.get('ENSEMBLE', price)

                    signal_emoji = {'BUY': '📈', 'SELL': '📉', 'HOLD': '➡️'}.get(signal, '➡️')

                    print(f"{status} {inst:12} | ₹{price:>9,.2f} | "
                          f"{signal_emoji} {signal:4} | Pred: ₹{ensemble:,.2f}")
                else:
                    print(f"{status} {inst:12} | ₹{price:>9,.2f} | Waiting for prediction...")
            else:
                print(f"{status} {inst:12} | No data yet ({count}/{MIN_CANDLES} candles)")

        # Show average prediction time
        if self.prediction_times:
            avg_time = np.mean(self.prediction_times[-10:])  # Last 10
            print(f"\n⚡ Avg prediction time: {avg_time:.3f}s")

        self.last_status_time = datetime.now()

    def _autosave(self):
        """Auto-save predictions"""
        if self.predictions:
            df = pd.DataFrame(self.predictions)
            filename = f'predictions_autosave_{datetime.now().strftime("%H%M%S")}.csv'
            df.to_csv(filename, index=False)
            print(f"💾 Auto-saved {len(self.predictions)} predictions to {filename}")
            self.last_autosave_time = datetime.now()

    def _print_final_summary(self):
        """Print final summary"""
        elapsed = (datetime.now() - self.start_time).total_seconds()

        print(f"\n📊 FINAL SUMMARY:")
        print(f"   Total runtime: {int(elapsed//60)}m {int(elapsed%60)}s")
        print(f"   Messages received: {self.message_count}")
        print(f"   Predictions generated: {self.prediction_count}")

        if self.predictions:
            df = pd.DataFrame(self.predictions)

            print(f"\n📈 Signal Distribution:")
            signal_counts = df['signal'].value_counts()
            for signal, count in signal_counts.items():
                emoji = {'BUY': '📈', 'SELL': '📉', 'HOLD': '➡️'}.get(signal, '➡️')
                pct = (count / len(df)) * 100
                print(f"   {emoji} {signal}: {count} ({pct:.1f}%)")

            # Save final predictions - MULTIPLE CSV FILES
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

            # 1. Detailed predictions
            detailed_file = f'predictions_detailed_{timestamp}.csv'
            df_detailed = df.copy()
            df_detailed['date'] = pd.to_datetime(df_detailed['time']).dt.date
            df_detailed['time_only'] = pd.to_datetime(df_detailed['time']).dt.time
            df_detailed.to_csv(detailed_file, index=False)
            print(f"\n💾 Detailed predictions: {detailed_file}")

            # 2. Summary by instrument
            summary_data = []
            for inst in df['instrument'].unique():
                inst_df = df[df['instrument'] == inst].sort_values('time')
                if len(inst_df) > 1:
                    summary_data.append({
                        'instrument': inst,
                        'total_predictions': len(inst_df),
                        'first_price': inst_df['price'].iloc[0],
                        'last_price': inst_df['price'].iloc[-1],
                        'price_change_pct': ((inst_df['price'].iloc[-1] - inst_df['price'].iloc[0]) /
                                           inst_df['price'].iloc[0]) * 100,
                        'buy_signals': len(inst_df[inst_df['signal'] == 'BUY']),
                        'sell_signals': len(inst_df[inst_df['signal'] == 'SELL']),
                        'hold_signals': len(inst_df[inst_df['signal'] == 'HOLD']),
                        'avg_rsi': inst_df['RSI'].mean() if 'RSI' in inst_df.columns else None
                    })

            summary_file = f'predictions_summary_{timestamp}.csv'
            pd.DataFrame(summary_data).to_csv(summary_file, index=False)
            print(f"💾 Summary: {summary_file}")

            # 3. Signals only
            signals_df = df[df['signal'].isin(['BUY', 'SELL'])][['time', 'instrument', 'signal', 'price']]
            signals_file = f'signals_{timestamp}.csv'
            signals_df.to_csv(signals_file, index=False)
            print(f"💾 Trading signals: {signals_file}")

            # Show performance by instrument
            print(f"\n💹 Price Movement:")
            for inst in df['instrument'].unique():
                inst_df = df[df['instrument'] == inst].sort_values('time')
                if len(inst_df) > 1:
                    first = inst_df['price'].iloc[0]
                    last = inst_df['price'].iloc[-1]
                    change = ((last - first) / first) * 100
                    print(f"   {inst}: {change:+.2f}%")

        print("\n" + "="*80)
        print("✅ All CSV files ready for visualization!")
        print("="*80)

    def _update_data(self, feed_data):
        """Update data from feed"""
        for inst_key, inst_data in feed_data.get('feeds', {}).items():
            inst_name = None
            for name, key in INSTRUMENTS.items():
                if key == inst_key:
                    inst_name = name
                    break

            if not inst_name:
                continue

            try:
                ltpc = inst_data['fullFeed']['indexFF']['ltpc']
                ohlc = inst_data['fullFeed']['indexFF']['marketOHLC']['ohlc']

                for candle in ohlc:
                    if candle['interval'] == '1d':
                        self.data[inst_name].append({
                            'time': datetime.fromtimestamp(int(ltpc['ltt'])/1000),
                            'open': float(candle['open']),
                            'high': float(candle['high']),
                            'low': float(candle['low']),
                            'close': float(candle['close']),
                            'ltp': float(ltpc['ltp'])
                        })
            except:
                pass

    async def _subscribe(self, websocket):
        """Subscribe to instruments"""
        await websocket.send(json.dumps({
            "guid": "continuous_trading",
            "method": "sub",
            "data": {
                "mode": "full",
                "instrumentKeys": list(INSTRUMENTS.values())
            }
        }).encode('utf-8'))
        await asyncio.sleep(1)

    def _get_auth(self):
        """Get WebSocket URL"""
        try:
            resp = requests.get(
                'https://api.upstox.com/v3/feed/market-data-feed/authorize',
                headers={
                    'Accept': 'application/json',
                    'Authorization': f'Bearer {ACCESS_TOKEN}'
                }
            )
            data = resp.json()

            if 'data' in data and 'authorized_redirect_uri' in data['data']:
                return data['data']['authorized_redirect_uri']
            else:
                print(f"Auth response: {data}")
                return None
        except Exception as e:
            print(f"Auth error: {e}")
            return None

    def _decode(self, msg):
        """Decode protobuf"""
        try:
            import market_feeder_pb2 as pb
            from google.protobuf.json_format import MessageToDict

            feed = pb.FeedResponse()
            feed.ParseFromString(msg)
            return MessageToDict(feed)
        except:
            return {}

# ============================================================================
# USAGE
# ============================================================================

if __name__ == "__main__":
    import nest_asyncio
    nest_asyncio.apply()

    print("""
╔════════════════════════════════════════════════════════════════════════════╗
║              CONTINUOUS PREDICTION SYSTEM                                  ║
║                                                                            ║
║  • Predictions run automatically every 10 seconds                         ║
║  • Status updates every 30 seconds                                        ║
║  • Auto-saves every 5 minutes                                             ║
║  • Real-time signal generation                                            ║
╚════════════════════════════════════════════════════════════════════════════╝

⚠️  Make sure to update ACCESS_TOKEN before running!

""")

    system = ContinuousTradingSystem()

    try:
        # Run for 60 minutes (adjust as needed)
        asyncio.run(system.run(duration_minutes=60))
    except KeyboardInterrupt:
        print("\n\n⚠️ Interrupted by user")
        system._print_final_summary()
    except Exception as e:
        print(f"\n❌ Error: {e}")
        import traceback
        traceback.print_exc()
```

```python
# === Auto-visualization pipeline for research-paper style figures ===
# Requirements: pandas, numpy, matplotlib, seaborn, networkx, sklearn
# install if needed: !pip install pandas numpy matplotlib seaborn networkx scikit-learn


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from math import pi
from sklearn.metrics import precision_score, recall_score
from google.colab import files  # <--- for downloading files
import os, shutil

# ---- Helper: attempt to find your df variable if not already named df ----
try:
    df  # if exists, fine
except NameError:
    # try common alternatives
    for cand in ["predictions_df","final_df","result_df","results_df","output_df","data"]:
        if cand in globals():
            df = globals()[cand]
            print(f"Auto-selected DataFrame variable: {cand}")
            break
    else:
        raise NameError("Please place your table into a pandas DataFrame named `df` and re-run this cell.")


# ---- Normalize column names (lowercase keys) and required columns ----
df = df.rename(columns=lambda c: str(c).strip())
cols = {c.lower():c for c in df.columns}
# map expected names
time_col = cols.get('time_only', cols.get('time', next(iter(cols.values()))))
date_col = cols.get('date', None)
price_col = cols.get('price', cols.get('close', None))
signal_col = cols.get('signal', None)
rsi_col = cols.get('rsi', None)
sma10_col = cols.get('sma_10', cols.get('sma10', None))
sma20_col = cols.get('sma_20', cols.get('sma20', None))
instrument_col = cols.get('instrument', cols.get('ticker', None))


# basic validation
if price_col is None or signal_col is None or instrument_col is None:
    raise ValueError("DataFrame must contain at least: price, signal, and instrument columns (names like 'price','signal','instrument').")


# ---- Parse timestamp: many of your times look like '26:30.6' (MM:SS.s) + date like '10/15/2025' ----
def parse_time_row(row):
    t = str(row.get(time_col,""))
    date = str(row.get(date_col,"")) if date_col else ""
    # try MMM:SS.s or H:M:S formats
    try:
        parts = t.split(':')
        if len(parts) == 2:
            minutes = int(parts[0])
            seconds = float(parts[1])
            td = pd.to_timedelta(minutes, unit='m') + pd.to_timedelta(seconds, unit='s')
            if date:
                return pd.to_datetime(date, dayfirst=False) + td
            else:
                return pd.Timestamp("1970-01-01") + td
        else:
            # fallback: direct parse
            return pd.to_datetime(date + " " + t)
    except Exception:
        try:
            return pd.to_datetime(date + " " + t)
        except Exception:
            return pd.NaT


df = df.copy()
df['__ts'] = df.apply(parse_time_row, axis=1)
# if parsing failed massively, fallback to index order
if df['__ts'].isna().sum() > len(df)*0.5:
    df['__ts'] = pd.date_range(start=pd.Timestamp("2025-10-15"), periods=len(df), freq='S')


df = df.sort_values('__ts').reset_index(drop=True)


# convert price to numeric
df[price_col] = pd.to_numeric(df[price_col], errors='coerce')


# indicator columns may exist; ensure numeric
for col in (rsi_col, sma10_col, sma20_col):
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')


# ---- Create direction labels from signal and actual move (next-tick) ----
# signal_dir: BUY->1 SELL->-1 HOLD->0
def map_signal(x):
    if pd.isna(x): return 0
    x = str(x).strip().upper()
    if 'BUY' in x: return 1
    if 'SELL' in x: return -1
    return 0


df['signal_dir'] = df[signal_col].apply(map_signal)


# actual direction: compare next observed price per instrument
df['next_price'] = df.groupby(instrument_col)[price_col].shift(-1)
df['actual_dir'] = np.sign(df['next_price'] - df[price_col])
df['actual_dir'] = df['actual_dir'].replace({0: -1})  # treat flat as non-up (so signal BUY won't be counted as correct)
df['actual_dir'] = df['actual_dir'].fillna(0)


# filter rows where actual_dir is defined
eval_df = df[df['actual_dir'] != 0].copy()


# compute per-instrument directional metrics (only consider BUY/SELL predictions, skip HOLDs)
metrics = []
for inst, sub in eval_df.groupby(instrument_col):
    preds = sub['signal_dir']
    trues = sub['actual_dir']
    # keep only non-hold predictions
    mask = preds != 0
    if mask.sum() == 0:
        acc = np.nan
        prec = np.nan
        rec = np.nan
    else:
        preds_masked = preds[mask]
        trues_masked = trues[mask]
        # map -1 -> 0, +1 -> 1 for precision/recall metrics
        y_pred = (preds_masked.values == 1).astype(int)
        y_true = (trues_masked.values == 1).astype(int)
        acc = (preds_masked.values == trues_masked.values).mean()
        try:
            prec = precision_score(y_true, y_pred, zero_division=0)
            rec = recall_score(y_true, y_pred, zero_division=0)
        except Exception:
            prec = np.nan; rec = np.nan
    metrics.append({'instrument': inst, 'directional_accuracy': acc, 'precision_up': prec, 'recall_up': rec, 'n_samples': int(mask.sum())})


metrics_df = pd.DataFrame(metrics).sort_values('instrument')


# === Styling for publication-like charts ===
plt.rcParams.update({
    'figure.dpi': 150,
    'font.size': 10,
    'axes.titlesize': 11,
    'axes.labelsize': 10,
    'legend.fontsize': 9
})
sns.set_style("whitegrid")


# ---------------- FIG 1: Biological Neuron (simple diagram) ----------------
fig, ax = plt.subplots(figsize=(6,3.5), dpi=800)
ax.add_patch(plt.Circle((0.5,0.5), 0.1))
ax.text(0.12, 0.65, "Dendrites", fontsize=11)
ax.add_patch(plt.Circle((0.4,0.5), 0.12, color='lightgrey', ec='k'))
ax.text(0.4, 0.5, "Cell\nBody", ha='center', va='center')
ax.annotate('', xy=(0.6, 0.5), xytext=(0.52,0.5), arrowprops=dict(arrowstyle='->', lw=2))
ax.text(0.7, 0.52, 'Axon', fontsize=11)
ax.set_title("Figure 1. Biological Neuron (schematic)")
ax.axis('off')
plt.tight_layout()
fig.savefig("fig1_bio_neuron.png", dpi=800, bbox_inches="tight")
plt.show()


# ---------------- FIG 2: Artificial Neuron ----------------
G = nx.DiGraph()
inputs = ['x1','x2','x3']
for i in inputs:
    G.add_edge(i, 'Σ')
G.add_edge('Σ', 'f(net)')
G.add_edge('f(net)', 'y')


plt.figure(figsize=(6,3.5), dpi=800)
pos = {'x1':(0,0.8),'x2':(0,0.5),'x3':(0,0.2),'Σ':(0.5,0.5),'f(net)':(0.75,0.5),'y':(1,0.5)}
nx.draw(G, pos, with_labels=True, node_size=1700, node_color="#d4f0ff", arrowsize=12)
plt.title("Figure 2. Artificial Neuron (weighted sum → activation → output)")
plt.axis('off')
plt.tight_layout()
plt.savefig("fig2_artificial_neuron.png", dpi=800, bbox_inches="tight")
plt.show()


# ---------------- FIG 3: Small ANN Architecture ----------------
G2 = nx.DiGraph()
inputs = ['x1','x2','x3']
hidden = ['h1','h2','h3']
out = ['y']
for i in inputs:
    for h in hidden:
        G2.add_edge(i,h)
for h in hidden:
    G2.add_edge(h,'y')


plt.figure(figsize=(6,3.5), dpi=800)
pos = {'x1':(0,0.8),'x2':(0,0.5),'x3':(0,0.2),'h1':(0.5,0.8),'h2':(0.5,0.5),'h3':(0.5,0.2),'y':(1,0.5)}
nx.draw(G2,pos,with_labels=True,node_size=1400,node_color="#eeffd4",arrowsize=10)
plt.title("Figure 3. ANN Architecture (Input → Hidden → Output)")
plt.axis('off')
plt.tight_layout()
plt.savefig("fig3_ann_arch.png", dpi=800, bbox_inches="tight")
plt.show()


# ---------------- FIG 4: Model-Performance Barplot (Directional accuracy by instrument) ----------------
plt.figure(figsize=(7,3.6), dpi=800)
sns.barplot(data=metrics_df, x='instrument', y='directional_accuracy', palette='Blues_d')
plt.ylim(0,1)
plt.ylabel('Directional Accuracy (BUY/SELL hit rate)')
plt.title("Figure 4. Directional Accuracy by Instrument (signal vs next tick movement)")
for i,row in metrics_df.iterrows():
    plt.text(i, row['directional_accuracy']+0.03, f"{row['directional_accuracy']:.2f}", ha='center', fontsize=9)
plt.tight_layout()
plt.savefig("fig4_directional_accuracy.png", dpi=800, bbox_inches="tight")
plt.show()


# ---------------- FIG 5: Price time series with BUY/SELL markers for one instrument (choose first) ----------------
inst0 = df[instrument_col].unique()[0]
sub = df[df[instrument_col]==inst0].copy()
plt.figure(figsize=(10,4), dpi=800)
plt.plot(sub['__ts'], sub[price_col], linewidth=1.2, label=f'{inst0} Price')
# markers
buy_mask = sub['signal_dir']==1
sell_mask = sub['signal_dir']==-1
plt.scatter(sub.loc[buy_mask,'__ts'], sub.loc[buy_mask, price_col], marker='^', s=40, label='BUY signal', zorder=3)
plt.scatter(sub.loc[sell_mask,'__ts'], sub.loc[sell_mask, price_col], marker='v', s=40, label='SELL signal', zorder=3)
plt.title(f"Figure 5. Price and Signals — {inst0}")
plt.xlabel("Timestamp")
plt.ylabel("Price")
plt.legend()
plt.tight_layout()
plt.savefig("fig5_price_signals.png", dpi=800, bbox_inches="tight")
plt.show()


# ---------------- FIG 6: Price + RSI + SMAs (if available) ----------------
fig, axes = plt.subplots(3,1,figsize=(10,7), sharex=True, dpi=800)
axes[0].plot(sub['__ts'], sub[price_col], lw=1.1)
axes[0].set_ylabel('Price')
axes[0].set_title(f'Figure 6. Price + RSI + SMAs — {inst0}')
if sma10_col in sub.columns:
    axes[0].plot(sub['__ts'], sub[sma10_col], lw=0.9, ls='--', label='SMA10')
if sma20_col in sub.columns:
    axes[0].plot(sub['__ts'], sub[sma20_col], lw=0.9, ls='-.', label='SMA20')
axes[0].legend()


if rsi_col in sub.columns:
    axes[1].plot(sub['__ts'], sub[rsi_col], lw=0.9)
    axes[1].axhline(70, ls='--', lw=0.7)
    axes[1].axhline(30, ls='--', lw=0.7)
    axes[1].set_ylabel('RSI')


# compute MACD if not present (fast=12, slow=26, signal=9) but on intraday short series it might be noisy
if 'MACD' in sub.columns and 'Signal' in sub.columns:
    axes[2].plot(sub['__ts'], sub['MACD'], label='MACD')
    axes[2].plot(sub['__ts'], sub['Signal'], label='Signal', ls='--')
    axes[2].legend()
else:
    # compute a simple MACD-like series with EMA
    price = sub[price_col].ffill()
    ema12 = price.ewm(span=12, adjust=False).mean()
    ema26 = price.ewm(span=26, adjust=False).mean()
    macd = ema12 - ema26
    signal_line = macd.ewm(span=9, adjust=False).mean()
    axes[2].plot(sub['__ts'], macd, label='MACD (est)')
    axes[2].plot(sub['__ts'], signal_line, label='Signal (est)', ls='--')
    axes[2].legend()
axes[2].set_ylabel('MACD')


plt.tight_layout()
fig.savefig("fig6_price_rsi_sma_macd.png", dpi=800, bbox_inches="tight")
plt.show()


# ---------------- FIG 7: Correlation heatmap of indicators (RSI, SMA10-SMA20 diff) ----------------
ind_cols = []
if rsi_col in df.columns: ind_cols.append(rsi_col)
if sma10_col in df.columns and sma20_col in df.columns:
    df['sma_diff'] = df[sma10_col] - df[sma20_col]
    ind_cols.append('sma_diff')
if len(ind_cols) >= 2:
    corr = df[ind_cols].corr()
    plt.figure(figsize=(4,3.5), dpi=800)
    sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title("Figure 7. Indicator Correlation")
    plt.tight_layout()
    plt.savefig("fig7_indicator_corr.png", dpi=800, bbox_inches="tight")
    plt.show()
else:
    print("Not enough indicator columns for correlation heatmap (need at least 2).")


# ---------------- FIG 8: Radar Chart comparing instruments (accuracy, precision, recall) ----------------
radar_df = metrics_df.set_index('instrument').fillna(0)
# pick metrics to show
radar_metrics = ['directional_accuracy','precision_up','recall_up']
# if missing columns, adapt
radar_metrics = [c for c in radar_metrics if c in radar_df.columns]
if len(radar_metrics) == 0:
    print("No metrics available for radar chart.")
else:
    categories = radar_metrics
    N = len(categories)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
    plt.figure(figsize=(6,5))
    ax = plt.subplot(111, polar=True)
    for idx, row in radar_df.iterrows():
        values = row[categories].values.flatten().tolist()
        values += values[:1]
        ax.plot(angles, values, linewidth=1.5, label=idx)
        ax.fill(angles, values, alpha=0.1)
    ax.set_thetagrids([a * 180/pi for a in angles[:-1]], categories)
    plt.title("Figure 8. Instrument Comparison (radar)")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.savefig("fig8_radar_instruments.png", dpi=800, bbox_inches="tight")
    plt.show()


# ---------------- Summary printout ----------------
print("\n=== Quick metrics summary ===")
display(metrics_df.style.background_gradient(cmap='Blues'))


# ---------------- ZIP AND DOWNLOAD ALL FIGURES ----------------
os.makedirs("paper_figures", exist_ok=True)
for f in os.listdir():
    if f.startswith("fig") and f.endswith(".png"):
        shutil.move(f, os.path.join("paper_figures", f))

shutil.make_archive("paper_figures", "zip", "paper_figures")
files.download("paper_figures.zip")  # will prompt download in browser
```

```python
# === Auto-visualization pipeline (research-paper style, HSV + Helvetica bold) ===
# Requirements: pandas, numpy, matplotlib, seaborn, networkx, sklearn
# install if needed: !pip install pandas numpy matplotlib seaborn networkx scikit-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
from math import pi
from sklearn.metrics import precision_score, recall_score
import matplotlib as mpl
import matplotlib.colors as mcolors

# ---- Helper: auto-detect DataFrame variable ----
try:
    df
except NameError:
    for cand in ["predictions_df","final_df","result_df","results_df","output_df","data"]:
        if cand in globals():
            df = globals()[cand]
            print(f"Auto-selected DataFrame variable: {cand}")
            break
    else:
        raise NameError("Please assign your data table to a DataFrame named `df`.")

# ---- Normalize column names ----
df = df.rename(columns=lambda c: str(c).strip())
cols = {c.lower(): c for c in df.columns}
time_col = cols.get('time_only', cols.get('time', next(iter(cols.values()))))
date_col = cols.get('date', None)
price_col = cols.get('price', cols.get('close', None))
signal_col = cols.get('signal', None)
rsi_col = cols.get('rsi', None)
sma10_col = cols.get('sma_10', cols.get('sma10', None))
sma20_col = cols.get('sma_20', cols.get('sma20', None))
instrument_col = cols.get('instrument', cols.get('ticker', None))

if price_col is None or signal_col is None or instrument_col is None:
    raise ValueError("Need at least 'price', 'signal', 'instrument' columns.")

# ---- Parse timestamp ----
def parse_time_row(row):
    t = str(row.get(time_col, ""))
    date = str(row.get(date_col, "")) if date_col else ""
    try:
        parts = t.split(':')
        if len(parts) == 2:
            minutes = int(parts[0]); seconds = float(parts[1])
            td = pd.to_timedelta(minutes, unit='m') + pd.to_timedelta(seconds, unit='s')
            if date:
                return pd.to_datetime(date) + td
            else:
                return pd.Timestamp("1970-01-01") + td
        else:
            return pd.to_datetime(date + " " + t)
    except Exception:
        try:
            return pd.to_datetime(date + " " + t)
        except Exception:
            return pd.NaT

df = df.copy()
df['__ts'] = df.apply(parse_time_row, axis=1)
if df['__ts'].isna().sum() > len(df)*0.5:
    df['__ts'] = pd.date_range(start=pd.Timestamp("2025-10-15"), periods=len(df), freq='S')

df = df.sort_values('__ts').reset_index(drop=True)
df[price_col] = pd.to_numeric(df[price_col], errors='coerce')
for col in (rsi_col, sma10_col, sma20_col):
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# ---- Directional metrics ----
def map_signal(x):
    if pd.isna(x): return 0
    x = str(x).upper()
    if 'BUY' in x: return 1
    if 'SELL' in x: return -1
    return 0

df['signal_dir'] = df[signal_col].apply(map_signal)
df['next_price'] = df.groupby(instrument_col)[price_col].shift(-1)
df['actual_dir'] = np.sign(df['next_price'] - df[price_col])
df['actual_dir'] = df['actual_dir'].replace({0: -1}).fillna(0)
eval_df = df[df['actual_dir'] != 0].copy()

metrics = []
for inst, sub in eval_df.groupby(instrument_col):
    preds = sub['signal_dir']; trues = sub['actual_dir']
    mask = preds != 0
    if mask.sum() == 0:
        acc = prec = rec = np.nan
    else:
        preds_masked = preds[mask]; trues_masked = trues[mask]
        y_pred = (preds_masked.values == 1).astype(int)
        y_true = (trues_masked.values == 1).astype(int)
        acc = (preds_masked.values == trues_masked.values).mean()
        prec = precision_score(y_true, y_pred, zero_division=0)
        rec = recall_score(y_true, y_pred, zero_division=0)
    metrics.append({
        'instrument': inst,
        'directional_accuracy': acc,
        'precision_up': prec,
        'recall_up': rec,
        'n_samples': int(mask.sum())
})
metrics_df = pd.DataFrame(metrics).sort_values('instrument')

# ---- HSV + Helvetica Styling ----
def hsv_to_rgb(h, s, v):  # h=0-360
    return mcolors.hsv_to_rgb((h/360, s, v))

palette_hsv = {
    "deep_blue": hsv_to_rgb(210, 0.9, 0.5),
    "dark_red": hsv_to_rgb(0, 0.9, 0.55),
    "dark_green": hsv_to_rgb(140, 0.9, 0.5),
    "dark_orange": hsv_to_rgb(30, 0.9, 0.6),
    "dark_purple": hsv_to_rgb(275, 0.8, 0.55)
}

mpl.rcParams.update({
    'font.family': 'Helvetica',
    'font.weight': 'bold',
    'axes.labelweight': 'bold',
    'axes.titlesize': 12,
    'axes.labelsize': 11,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'axes.linewidth': 1.8,
    'grid.linewidth': 0.6,
    'pdf.fonttype': 42,
    'ps.fonttype': 42
})
sns.set_style("whitegrid")

# ---- FIGURES ----

# FIG 1: Biological Neuron
fig, ax = plt.subplots(figsize=(6,3.5))
ax.add_patch(plt.Circle((0.4,0.5), 0.12, color=palette_hsv["dark_green"], ec='k', lw=2))
ax.text(0.12, 0.65, "Dendrites", fontsize=11, weight='bold')
ax.text(0.4, 0.5, "Cell\nBody", ha='center', va='center', fontsize=10, weight='bold')
ax.annotate('', xy=(0.6,0.5), xytext=(0.52,0.5), arrowprops=dict(arrowstyle='->', lw=2, color='k'))
ax.text(0.7, 0.52, 'Axon', fontsize=11, weight='bold')
ax.set_title("Figure 1. Biological Neuron (schematic)", weight='bold')
ax.axis('off')
plt.tight_layout(); plt.show()

# FIG 2: Artificial Neuron
G = nx.DiGraph()
inputs = ['x1','x2','x3']
for i in inputs: G.add_edge(i, 'Σ')
G.add_edge('Σ','f(net)'); G.add_edge('f(net)','y')
plt.figure(figsize=(6,3.5))
pos = {'x1':(0,0.8),'x2':(0,0.5),'x3':(0,0.2),'Σ':(0.5,0.5),'f(net)':(0.75,0.5),'y':(1,0.5)}
nx.draw(G,pos,with_labels=True,node_size=1700,node_color="#f0e68c",
        arrowsize=14,width=2,font_weight='bold',font_family='Helvetica')
plt.title("Figure 2. Artificial Neuron (weighted sum → activation → output)", weight='bold')
plt.axis('off'); plt.tight_layout(); plt.show()

# FIG 3: Small ANN
G2 = nx.DiGraph()
inputs = ['x1','x2','x3']; hidden=['h1','h2','h3']; out=['y']
for i in inputs:
    for h in hidden: G2.add_edge(i,h)
for h in hidden: G2.add_edge(h,'y')
plt.figure(figsize=(6,3.5))
pos = {'x1':(0,0.8),'x2':(0,0.5),'x3':(0,0.2),
       'h1':(0.5,0.8),'h2':(0.5,0.5),'h3':(0.5,0.2),'y':(1,0.5)}
nx.draw(G2,pos,with_labels=True,node_size=1400,node_color="#d4f0ff",
        width=2,arrowsize=12,font_weight='bold',font_family='Helvetica')
plt.title("Figure 3. ANN Architecture (Input → Hidden → Output)", weight='bold')
plt.axis('off'); plt.tight_layout(); plt.show()

# FIG 4: Directional Accuracy Barplot
plt.figure(figsize=(7,3.6))
sns.barplot(data=metrics_df, x='instrument', y='directional_accuracy',
            palette=[palette_hsv["deep_blue"]])
plt.ylim(0,1); plt.ylabel('Directional Accuracy', weight='bold')
plt.title("Figure 4. Directional Accuracy by Instrument", weight='bold')
for i,row in metrics_df.iterrows():
    plt.text(i, row['directional_accuracy']+0.03, f"{row['directional_accuracy']:.2f}",
             ha='center', fontsize=9, weight='bold', color='black')
sns.despine(); plt.tight_layout(); plt.show()

# FIG 5: Price time series with BUY/SELL
inst0 = df[instrument_col].unique()[0]
sub = df[df[instrument_col]==inst0].copy()
plt.figure(figsize=(10,4))
plt.plot(sub['__ts'], sub[price_col], color=palette_hsv["deep_blue"],
         linewidth=2.0, label=f'{inst0} Price')
buy_mask = sub['signal_dir']==1
sell_mask = sub['signal_dir']==-1
plt.scatter(sub.loc[buy_mask,'__ts'], sub.loc[buy_mask, price_col],
            marker='^', s=55, color=palette_hsv["dark_green"], edgecolors='k', label='BUY')
plt.scatter(sub.loc[sell_mask,'__ts'], sub.loc[sell_mask, price_col],
            marker='v', s=55, color=palette_hsv["dark_red"], edgecolors='k', label='SELL')
plt.title(f"Figure 5. Price and Signals — {inst0}", weight='bold')
plt.xlabel("Timestamp", weight='bold'); plt.ylabel("Price", weight='bold')
plt.legend(frameon=True); sns.despine(); plt.tight_layout(); plt.show()

# FIG 6: Price + RSI + SMAs
fig, axes = plt.subplots(3,1,figsize=(10,7), sharex=True)
axes[0].plot(sub['__ts'], sub[price_col], lw=1.8, color=palette_hsv["deep_blue"], label='Price')
if sma10_col in sub.columns:
    axes[0].plot(sub['__ts'], sub[sma10_col], lw=1.4, ls='--', color=palette_hsv["dark_green"], label='SMA10')
if sma20_col in sub.columns:
    axes[0].plot(sub['__ts'], sub[sma20_col], lw=1.4, ls='-.', color=palette_hsv["dark_orange"], label='SMA20')
axes[0].legend(); axes[0].set_ylabel('Price', weight='bold')
axes[0].set_title(f'Figure 6. Price + RSI + SMAs — {inst0}', weight='bold')

if rsi_col in sub.columns:
    axes[1].plot(sub['__ts'], sub[rsi_col], lw=1.3, color=palette_hsv["dark_purple"])
    axes[1].axhline(70, ls='--', lw=1.0, color='grey')
    axes[1].axhline(30, ls='--', lw=1.0, color='grey')
    axes[1].set_ylabel('RSI', weight='bold')

price = sub[price_col].ffill()
ema12 = price.ewm(span=12, adjust=False).mean()
ema26 = price.ewm(span=26, adjust=False).mean()
macd = ema12 - ema26
signal_line = macd.ewm(span=9, adjust=False).mean()
axes[2].plot(sub['__ts'], macd, color=palette_hsv["dark_blue"] if "dark_blue" in palette_hsv else palette_hsv["deep_blue"], lw=1.5, label='MACD')
axes[2].plot(sub['__ts'], signal_line, color=palette_hsv["dark_red"], lw=1.5, ls='--', label='Signal')
axes[2].legend(); axes[2].set_ylabel('MACD', weight='bold')
plt.tight_layout(); plt.show()

# FIG 7: Correlation heatmap
ind_cols = []
if rsi_col in df.columns: ind_cols.append(rsi_col)
if sma10_col in df.columns and sma20_col in df.columns:
    df['sma_diff'] = df[sma10_col] - df[sma20_col]; ind_cols.append('sma_diff')
if len(ind_cols) >= 2:
    corr = df[ind_cols].corr()
    plt.figure(figsize=(4,3.5))
    sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1,
                annot_kws={"weight":'bold'})
    plt.title("Figure 7. Indicator Correlation", weight='bold')
    plt.tight_layout(); plt.show()
else:
    print("Not enough indicator columns for correlation heatmap.")

# FIG 8: Radar chart
radar_df = metrics_df.set_index('instrument').fillna(0)
radar_metrics = [c for c in ['directional_accuracy','precision_up','recall_up'] if c in radar_df.columns]
if len(radar_metrics) > 1:
    categories = radar_metrics; N = len(categories)
    angles = [n / float(N) * 2 * pi for n in range(N)] + [0]
    plt.figure(figsize=(6,5))
    ax = plt.subplot(111, polar=True)
    for idx, row in radar_df.iterrows():
        values = row[categories].tolist() + [row[categories[0]]]
        ax.plot(angles, values, linewidth=2, label=idx)
        ax.fill(angles, values, alpha=0.1)
    ax.set_thetagrids([a * 180/pi for a in angles[:-1]], categories, weight='bold')
    plt.title("Figure 8. Instrument Comparison (radar)", weight='bold')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout(); plt.show()
else:
    print("Not enough metrics for radar chart.")

# ---- Summary ----
print("\n=== Quick metrics summary ===")
display(metrics_df.style.background_gradient(cmap='Blues'))
```

```python
# === Robust Heatmap Generator for Technical Indicators (Colab Safe) ===
# Automatically fixes 'price' column issue, installs dependencies, and handles missing indicators.

# Install TA library if not already
!pip install -q pandas_ta

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_ta as ta
from IPython.display import display

# ----------------------------------------------------------
# STEP 1: Load your data here
# ----------------------------------------------------------
# Example: df = pd.read_csv('/content/your_data.csv')
# Replace with your actual file name or DataFrame if already loaded
try:
    df
    print("✅ Existing dataframe detected.")
except NameError:
    print("⚠️ No dataframe named 'df' found. Please load your CSV below:")
    # Example placeholder (replace with your actual dataset)
    df = pd.read_csv('/content/your_data.csv')
    print("Loaded sample data:")
    display(df.head())

# ----------------------------------------------------------
# STEP 2: Fix column naming for compatibility
# ----------------------------------------------------------
# Convert all column names to lowercase for safety
df.columns = [c.lower() for c in df.columns]

# If your dataset uses 'price' instead of 'close', rename it
if 'price' in df.columns and 'close' not in df.columns:
    df = df.rename(columns={'price': 'close'})
    print("🔧 Renamed 'price' → 'close' for TA calculations.")

# If your dataset uses 'datetime', ensure it's sorted
if 'datetime' in df.columns:
    df = df.sort_values('datetime').reset_index(drop=True)

# ----------------------------------------------------------
# STEP 3: Compute or ensure indicators exist
# ----------------------------------------------------------
def ensure_indicators(df):
    df2 = df.copy()

    # RSI (14)
    if 'rsi' not in df2.columns:
        df2['rsi'] = ta.rsi(df2['close'], length=14)

    # MACD (12, 26, 9)
    macd = ta.macd(df2['close'], fast=12, slow=26, signal=9)
    df2['macd'] = macd.iloc[:, 0]
    df2['macd_signal'] = macd.iloc[:, 2]

    # SMAs
    for p in [10, 20, 50, 200]:
        name = f'sma_{p}'
        if name not in df2.columns:
            df2[name] = df2['close'].rolling(window=p, min_periods=1).mean()

    # Bollinger Bands (20, 2)
    bb = ta.bbands(df2['close'], length=20, std=2)
    if bb is not None and not bb.empty:
        df2['boll_upper'] = bb.iloc[:, 0]
        df2['boll_lower'] = bb.iloc[:, 2]
    else:
        df2['boll_upper'] = np.nan
        df2['boll_lower'] = np.nan

    # Derived metrics
    df2['sma_diff'] = df2['sma_10'] - df2['sma_20']
    df2['boll_width'] = df2['boll_upper'] - df2['boll_lower']

    # Fill NaN values
    df2 = df2.fillna(method='ffill').fillna(method='bfill')

    return df2

df = ensure_indicators(df)

# ----------------------------------------------------------
# STEP 4: Choose indicators for correlation
# ----------------------------------------------------------
indicators = [
    'close', 'volume', 'rsi', 'macd', 'macd_signal',
    'sma_10', 'sma_20', 'sma_50', 'sma_200',
    'boll_upper', 'boll_lower', 'sma_diff', 'boll_width'
]

# Filter only existing columns
existing = [col for col in indicators if col in df.columns]
print("\n📊 Using indicators for correlation heatmap:")
print(existing)

df_corr = df[existing].copy()

# ----------------------------------------------------------
# STEP 5: Compute and plot correlation heatmap
# ----------------------------------------------------------
corr = df_corr.corr(method='pearson')

plt.figure(figsize=(11, 9))
sns.set(font_scale=0.95)
ax = sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="vlag",
    center=0,
    linewidths=0.4,
    cbar_kws={"shrink": 0.6}
)
ax.set_title("Indicator Correlation Matrix", fontsize=14)
plt.tight_layout()
plt.show()

# ----------------------------------------------------------
# STEP 6: Save high-resolution heatmap image
# ----------------------------------------------------------
output_path = "/content/indicator_correlation_heatmap.png"
plt.savefig(output_path, dpi=300, bbox_inches="tight")
print(f"\n✅ Heatmap saved successfully at: {output_path}")

# Optional: Display correlation matrix
display(corr)
```

```python
# ================================
# MASTER CELL: compute_features + predict_live
# ================================

import numpy as np
import pandas as pd

# 1) Feature Engineering (same as training)
def compute_features(df):
    df = df.copy()
    s = df['close']

    # SMA
    df['sma10'] = s.rolling(10, min_periods=1).mean()
    df['sma20'] = s.rolling(20, min_periods=1).mean()

    # RSI 14
    delta = s.diff()
    up = delta.clip(lower=0)
    down = -delta.clip(upper=0)
    avg_up = up.rolling(14, min_periods=1).mean()
    avg_down = down.rolling(14, min_periods=1).mean()
    rs = avg_up / (avg_down + 1e-9)
    df['rsi14'] = 100 - (100 / (1 + rs))

    # MACD
    ema12 = s.ewm(span=12, adjust=False).mean()
    ema26 = s.ewm(span=26, adjust=False).mean()
    df['macd'] = ema12 - ema26
    df['macd_sig'] = df['macd'].ewm(span=9, adjust=False).mean()

    # Bollinger width
    ma20 = s.rolling(20, min_periods=1).mean()
    std20 = s.rolling(20, min_periods=1).std().fillna(0)
    df['boll_w'] = ((ma20 + 2*std20) - (ma20 - 2*std20)) / (ma20 + 1e-9)

    df = df.fillna(0)
    return df


# 2) Convert rolling window to model feature vector
def make_feature_vector(rolling_df):
    feats = compute_features(rolling_df)
    last = feats.iloc[-1]

    # feature_cols must already be loaded earlier
    global feature_cols

    used_cols = [col for col in feature_cols if col in feats.columns]
    x = np.array([last[col] for col in used_cols], dtype=float).reshape(1, -1)

    # scale if scaler exists
    global scaler
    if scaler is not None:
        x = scaler.transform(x)

    return x, last.to_dict(), used_cols


# 3) Predict live signal
def predict_live(rolling_df, threshold=0.6):
    x, feat_dump, used_cols = make_feature_vector(rolling_df)

    global model

    try:
        prob = float(model.predict_proba(x)[0][1])
    except:
        prob = float(model.predict(x)[0])

    signal = 1 if prob >= threshold else 0

    return {
        "prob": prob,
        "signal": signal,
        "features": feat_dump,
        "used_cols": used_cols
    }

print("Loaded: compute_features, make_feature_vector, predict_live")
```

```python
# ONE-SHOT: use existing `df` (Upstox OHLC) -> build features -> train tiny model -> save -> benchmark latency
import numpy as np, pandas as pd, time, gc, matplotlib.pyplot as plt, os
from pathlib import Path
print("Starting one-shot realtime pipeline using 'df' (Upstox data)...")

# 0) sanity
G = globals()
if 'df' not in G or not isinstance(G['df'], pd.DataFrame):
    raise RuntimeError("No DataFrame named 'df' found in globals. Re-run the Upstox fetch cell first.")
raw = G['df'].copy().reset_index(drop=True)
print("Found df shape:", raw.shape)

# 1) normalize column names (flatten possible multiindex-like names)
raw_cols = [str(c) for c in raw.columns]
raw.columns = raw_cols

# find Close column
close_col = None
for c in raw.columns:
    if c.lower().startswith('close') or c.lower()=='close' or 'close' in c.lower():
        close_col = c; break
if close_col is None:
    raise RuntimeError("Could not find a 'Close' column in df.")

# 2) build feature_df (minimal indicators used by prediction notebook)
feat = pd.DataFrame()
feat['Nifty_Close'] = pd.to_numeric(raw[close_col], errors='coerce')
# Volume if present (try common names)
vol_col = None
for c in raw.columns:
    if 'vol' in c.lower():
        vol_col = c; break
feat['Nifty_Volume'] = pd.to_numeric(raw[vol_col], errors='coerce') if vol_col is not None else 0.0

# basic indicators (fast)
feat['SMA_50'] = feat['Nifty_Close'].rolling(50, min_periods=1).mean()
feat['SMA_200'] = feat['Nifty_Close'].rolling(200, min_periods=1).mean()
ma20 = feat['Nifty_Close'].rolling(20, min_periods=1).mean()
sd20 = feat['Nifty_Close'].rolling(20, min_periods=1).std(ddof=0)
feat['Bollinger_Upper'] = ma20 + 2*sd20
feat['Bollinger_Lower'] = ma20 - 2*sd20
delta = feat['Nifty_Close'].diff()
up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)
roll_up = up.rolling(14, min_periods=1).mean()
roll_down = down.rolling(14, min_periods=1).mean()
rs = roll_up / (roll_down.replace(0, np.nan))
feat['RSI'] = 100 - (100/(1+rs)).fillna(50)
ema12 = feat['Nifty_Close'].ewm(span=12, adjust=False).mean()
ema26 = feat['Nifty_Close'].ewm(span=26, adjust=False).mean()
feat['MACD'] = ema12 - ema26
feat['Signal_Line'] = feat['MACD'].ewm(span=9, adjust=False).mean()
feat['ret_1'] = feat['Nifty_Close'].pct_change().fillna(0)

# keep same order / expected columns
cols_keep = ['Nifty_Close','Nifty_Volume','RSI','MACD','Signal_Line','SMA_50','SMA_200','Bollinger_Upper','Bollinger_Lower','ret_1']
for c in cols_keep:
    if c not in feat.columns:
        feat[c] = np.nan
feature_df = feat[cols_keep].copy().reset_index(drop=True)
G['feature_df'] = feature_df
print("Built feature_df shape:", feature_df.shape)

# 3) build sequences (look_back=30)
LOOK_BACK = 30
X_list, y_list = [], []
for i in range(LOOK_BACK, len(feature_df)):
    win = feature_df.iloc[i-LOOK_BACK:i].values  # (look_back, n_features)
    X_list.append(win)
    nxt = feature_df['Nifty_Close'].iloc[i]
    prev = feature_df['Nifty_Close'].iloc[i-1]
    y_list.append(1 if (nxt - prev) / prev > 0 else 0)
X_all = np.array(X_list); y_all = np.array(y_list)
n_samples = X_all.shape[0]
if n_samples < 40:
    raise RuntimeError(f"Too few samples ({n_samples}) after windowing. Need >=40. Current df length: {len(feature_df)}")
split = int(0.75 * n_samples)
X_train_3d = X_all[:split]; X_test_3d = X_all[split:]
y_train = y_all[:split]; y_test = y_all[split:]
G['X_train_3d']=X_train_3d; G['X_test_3d']=X_test_3d
G['y_train']=y_train; G['y_test']=y_test
print("Sequences built -> samples:", n_samples, "train:", X_train_3d.shape[0], "test:", X_test_3d.shape[0])

# 4) train a tiny BiLSTM (safe)
try:
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Bidirectional, LSTM, Dense, Dropout, InputLayer
    from tensorflow.keras.optimizers import Adam
    from tensorflow.keras.callbacks import EarlyStopping
except Exception as e:
    raise RuntimeError("Keras not available in this runtime: " + str(e))

look_back = X_train_3d.shape[1]; n_feat = X_train_3d.shape[2]
model = Sequential([InputLayer(input_shape=(look_back, n_feat)),
                    Bidirectional(LSTM(16, return_sequences=False)),
                    Dropout(0.2),
                    Dense(8, activation='relu'),
                    Dense(1, activation='sigmoid')])
model.compile(optimizer=Adam(1e-3), loss='binary_crossentropy', metrics=['accuracy'])
print("Model compiled. Training for a few epochs (low risk of OOM).")
es = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True, verbose=0)
hist = model.fit(X_train_3d, y_train, validation_data=(X_test_3d, y_test), epochs=12, batch_size=8, callbacks=[es], verbose=2)

# 5) save model & expose realtime_model
MODEL_FILE = "model.h5"
model.save(MODEL_FILE)
G['realtime_model'] = model
print("Saved model to", MODEL_FILE)

# 6) store predictions in model_preds_dict for compatibility
preds_test = model.predict(X_test_3d).ravel()
preds_train = model.predict(X_train_3d).ravel()
mpd = G.get('model_preds_dict', {})
mpd['trained_from_upstox__X_test'] = preds_test
mpd['trained_from_upstox__X_train'] = preds_train
G['model_preds_dict'] = mpd
print("Stored predictions keys:", list(mpd.keys()))

# 7) small latency benchmark (50 trials) - end2end & inference-only
def infer_only_time(m, X, warm=1):
    for _ in range(warm):
        try:
            _ = m.predict(X[:1])
        except Exception:
            pass
    t0 = time.perf_counter_ns(); _ = m.predict(X); t1 = time.perf_counter_ns()
    return (t1 - t0) / 1e6

def end2end_time(m, feature_df, look_back=LOOK_BACK):
    t0 = time.perf_counter_ns()
    X = feature_df.iloc[-look_back:].values.reshape(1, look_back, n_feat).astype('float32')
    t1 = time.perf_counter_ns()
    inf_ms = infer_only_time(m, X, warm=1)
    t2 = time.perf_counter_ns()
    return ((t2 - t0) / 1e6, inf_ms)

trials = 50
end2end = []; infs = []
for i in range(trials):
    t_total, t_inf = end2end_time(model, feature_df, look_back=LOOK_BACK)
    end2end.append(t_total); infs.append(t_inf)
end2end = np.array(end2end); infs = np.array(infs)

def stats(arr):
    return {'median_ms': float(np.median(arr)), 'mean_ms': float(np.mean(arr)),
            'p95_ms': float(np.percentile(arr,95)), 'p99_ms': float(np.percentile(arr,99)),
            'std_ms': float(np.std(arr))}

s_end = stats(end2end); s_inf = stats(infs)
print("Latency benchmark (50 trials) complete.")
print("END-TO-END stats (ms):", s_end)
print("INFERENCE-ONLY stats (ms):", s_inf)

# 8) save histograms at high resolution
os.makedirs('latency_outputs', exist_ok=True)
for arr, name in [(end2end, 'end2end'), (infs, 'inference')]:
    plt.figure(figsize=(6,3.5))
    plt.hist(arr, bins=50)
    plt.xlabel('ms'); plt.title(f'Latency {name}')
    plt.tight_layout()
    png = f"latency_outputs/latency_{name}.png"
    svg = f"latency_outputs/latency_{name}.svg"
    plt.savefig(png, dpi=800)
    plt.savefig(svg)
    plt.show()
    print("Saved:", png, svg)

# 9) expose summary
G['latency_stats'] = {'end2end': s_end, 'inf_only': s_inf}
G['latency_samples'] = {'end2end': end2end, 'inf': infs}
print("Saved latency stats to globals['latency_stats']. Done.")
gc.collect()
```

```python
# FINAL REAL-TIME PIPELINE (NO RETRAINING)
import json, time
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model

# 1. Load model
model = load_model("model.h5")
print("Model loaded")

# 2. Load feature columns
with open("feature_columns.json", "r") as f:
    FEATURE_COLS = json.load(f)

LOOK_BACK = 30  # safe to hardcode
print("Using LOOK_BACK =", LOOK_BACK)
print("Feature columns:", FEATURE_COLS)

# 3. Ensure Upstox df exists
if 'df' not in globals():
    raise RuntimeError("Upstox dataframe `df` not found. Run the Upstox fetch cell first.")

df_rt = df.copy().reset_index(drop=True)

# 4. Build real-time feature window
def build_realtime_window(df, feature_cols, look_back):
    missing = [c for c in feature_cols if c not in df.columns]
    if missing:
        raise RuntimeError(f"Missing required columns in df: {missing}")

    window = df[feature_cols].iloc[-look_back:].values
    return window.reshape(1, look_back, len(feature_cols)).astype("float32")

X_rt = build_realtime_window(df_rt, FEATURE_COLS, LOOK_BACK)
print("Realtime input shape:", X_rt.shape)

# 5. Measure inference latency
t0 = time.perf_counter_ns()
pred = model.predict(X_rt)
t1 = time.perf_counter_ns()

latency_ms = (t1 - t0) / 1e6

print("Prediction:", float(pred[0][0]))
print("Inference latency (ms):", latency_ms)
```
